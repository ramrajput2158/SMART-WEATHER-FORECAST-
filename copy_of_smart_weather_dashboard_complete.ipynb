# 🌦 Smart Weather Forecast Dashboard
Google Colab Project using **WeatherAPI + Gradio**.

Features:
- Current Location (IP based)
- Search any city
- Weather icon
- Temperature, humidity, wind, pressure
- Rain indication
- 5-day forecast

```python

import requests
import gradio as gr

def get_weather(city):
    query = city.strip() if city.strip() else "auto:ip"

    url=f"https://api.weatherapi.com/v1/forecast.json?key={API_KEY}&q={query}&days=5&aqi=yes&alerts=no"

    r=requests.get(url)

    if r.status_code!=200:
        return "API Error:\n"+r.text, None, ""

    data=r.json()

    c=data["current"]
    loc=data["location"]

    forecast=""
    for day in data["forecast"]["forecastday"]:
        forecast += (
            f"{day['date']} | "
            f"{day['day']['condition']['text']} | "
            f"{day['day']['maxtemp_c']}°C / "
            f"{day['day']['mintemp_c']}°C\n"
        )

    umbrella="YES ☂" if c["precip_mm"]>0 else "NO"

    report=f"""
📍 {loc['name']}, {loc['country']}

🌡 Temperature : {c['temp_c']} °C
🤒 Feels Like : {c['feelslike_c']} °C
☁ Condition : {c['condition']['text']}
💧 Humidity : {c['humidity']} %
🌬 Wind : {c['wind_kph']} km/h
🌡 Pressure : {c['pressure_mb']} mb
👁 Visibility : {c['vis_km']} km

🌧 Rain : {c['precip_mm']} mm
☂ Carry Umbrella : {umbrella}

========== 5 Day Forecast ==========
{forecast}
"""

    icon="https:"+c["condition"]["icon"]

    return report, icon, forecast

demo=gr.Interface(
    fn=get_weather,
    inputs=gr.Textbox(
        label="Enter City (Leave blank for Current Location)",
        placeholder="Delhi"
    ),
    outputs=[
        gr.Textbox(label="Weather Report", lines=18),
        gr.Image(label="Weather Icon"),
        gr.Textbox(label="5-Day Forecast", lines=6)
    ],
    title="🌦 Smart Weather Forecast Dashboard",
    description="Leave the city blank to use approximate current location (IP based)."
)

demo.launch(share=True)
```
